```python
# Initialize lists to store max and mean fitness values
max_fitness_values = []
mean_fitness_values = []

# Iterate through each DataFrame
for i, df in enumerate(generation_dfs):
    # Calculate max and mean fitness
    max_fitness = df['fitness'].max()
    mean_fitness = df['fitness'].mean()
    
    max_fitness_values.append(max_fitness)
    mean_fitness_values.append(mean_fitness)
    
    # Print for verification
    print(f"Generation {i} - Max Fitness: {max_fitness}, Mean Fitness: {mean_fitness}")

# Plotting
generations = list(range(len(generation_dfs)))

# Plot max fitness
plt.figure(figsize=(10, 6))
plt.plot(generations, max_fitness_values, marker='o', label='Max Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Max Fitness by Generation')
plt.legend()
plt.show()

# Plot mean fitness
plt.figure(figsize=(10, 6))
plt.plot(generations, mean_fitness_values, marker='o', color='r', label='Mean Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Mean Fitness by Generation')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of DataFrames
generation_dfs = [genetic_gen0, genetic_gen1, genetic_gen2, genetic_gen3, genetic_gen4, genetic_gen5, genetic_gen6, genetic_gen7, genetic_gen8, genetic_gen9, genetic_gen10]

# Create a figure and axes for the density plot
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through each subDataFrame and create a density plot for the fitness column
for i, df in enumerate(generation_dfs):
    label = f'Generation {i}'
    if i == 0 or i == 10:  # Check for Generation 0 and Generation 10
        sns.kdeplot(data=df, x='fitness', label=label, ax=ax, fill=True)  # Use fill=True for filling the plot
    else:
        sns.kdeplot(data=df, x='fitness', label=label, ax=ax)

ax.set_xlabel('Fitness')
ax.set_ylabel('Density')
ax.set_title('Density Plots for Fitness')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a figure and axes for the scatter plots
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through each subDataFrame and create scatter plots for the fitness column
for i, df in enumerate(generation_dfs):
    ax.scatter(df.index, df['fitness'], label=f'Generation {i}')

ax.set_xlabel('Index')
ax.set_ylabel('Fitness')
ax.set_title('Scatter Plots for Fitness')
ax.legend()
plt.tight_layout()
plt.show()
```
